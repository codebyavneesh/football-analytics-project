# Rating column ki missing values ko mean se fill karo, Birth_Date ko datetime format mein convert karo

```python
from sqlalchemy import create_engine, text
query="""SELECT * FROM fulldata"""

df = pd.read_sql(query, engine)

df['Birth_Date']=pd.to_datetime(df['Birth_Date'])

with engine.begin() as conn:
    conn.execute(text("DROP TABLE IF EXISTS fullData"))

# New updated table save karo
df.to_sql(
    name="fulldata",
    con=engine,
    index=False
)

df.info()
```
